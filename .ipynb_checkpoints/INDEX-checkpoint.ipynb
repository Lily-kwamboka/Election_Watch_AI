```python
#Importing Libraries
import os
os.environ["PROTOCOL_BUFFERS_PYTHON_IMPLEMENTATION"] = "python"
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import array_to_img, img_to_array, load_img
import numpy as np
import cv2
import random
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

# Data loading and preprocessing

```python
datagen = ImageDataGenerator(
    rescale=1./255,      
    validation_split=0.2 
)
# Training data
train_data = datagen.flow_from_directory(
    'hand_gestures/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)
#Validation
val_data = datagen.flow_from_directory(
    'hand_gestures/',
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

# Buiding the model

```python
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D(pool_size=(2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

    Flatten(),
    Dense(64, activation='relu'),
    Dense(train_data.num_classes, activation='softmax')  
])
```

# Compiling and Training the Model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_data, epochs=10, validation_data=val_data)
```

```python
#  training & validation accuracy
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
plt.show()
```

- Training Accuracy rises quickly and reaches 100%.

- Validation Accuracy is also very high and stays stable across epochs.

- There is no major overfitting, since validation accuracy doesn't drop off.

```python
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.grid(True)
plt.show()
```

- The model has learned to classify gestures very well on both training and validation data.

- The performance curve is smooth, which suggests good convergence.

```python
val_loss, val_accuracy = model.evaluate(val_data)
print(f"Validation Accuracy: {val_accuracy:.4f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

#  class labels
class_labels = list(val_data.class_indices.keys())

#images and labels from validation data
images, labels = next(val_data)

# Predict on these images
preds = model.predict(images)
pred_classes = np.argmax(preds, axis=1)
true_classes = np.argmax(labels, axis=1)

#  first 9 images with predictions
plt.figure(figsize=(12, 10))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i])
    plt.axis('off')
    plt.title(f"True: {class_labels[true_classes[i]]}\nPred: {class_labels[pred_classes[i]]}",
              color='green' if pred_classes[i] == true_classes[i] else 'red')
plt.tight_layout()
plt.show()
```

It is taking real images from validation data , making predictions, and showing correct vs. incorrect guesses using  trained model.

- We are appending it to a gesture buffer that remembers the last 5 predictions.

- If it detects the pattern:
'Holding_Ballot' → 'Hands_Down' → 'Open_Palm' ,
it logs a ballot drop and prevents duplicate logs using drop_logged.


```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load model and classes
model = load_model("gesture_model.keras")
class_labels = ['Two_Fingers_Edge', 'Hands_Down', 'Holding_Ballot', 'Moved_Palm', 'Open_Palm']

# For tracking recent gestures
gesture_buffer = deque(maxlen=5)
drop_logged = False

#Preprocessing
def preprocess_frame(frame):
    img = cv2.resize(frame, (64, 64)) 
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype("float32") / 255.0
    img = np.expand_dims(img, axis=0)
    return img

# Logic to detect if a ballot drop has occurred
def detect_ballot_drop(predicted_class):
    global drop_logged
    gesture_buffer.append(predicted_class)
    pattern = list(gesture_buffer)

    if 'Holding_Ballot' in pattern and 'Hands_Down' in pattern and 'Open_Palm' in pattern:
        if not drop_logged:
            print("Ballot drop detected!")
            with open("drop_log.txt", "a") as f:
                f.write(f"Drop detected at {time.ctime()}\n")
            drop_logged = True
    else:
        drop_logged = False

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Define ROI and show it
    roi = frame[100:300, 100:300] 
    cv2.rectangle(frame, (100, 100), (300, 300), (255, 0, 0), 2)

    img = cv2.resize(roi, (64, 64))  
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_rgb = img_rgb.astype("float32") / 255.0
    img_rgb = np.expand_dims(img_rgb, axis=0)

    # Predict
    prediction = model.predict(img_rgb, verbose=0)
    predicted_class = class_labels[np.argmax(prediction)]

   
    cv2.putText(frame, f"Gesture: {predicted_class}", (10, 40), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow("Gesture Detection", frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
